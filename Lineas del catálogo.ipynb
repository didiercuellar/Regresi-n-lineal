```python
from astropy.io import fits 
from astropy.table import Table 
import numpy as np 
```

```python
path=''
```

```python
hdul = fits.open('dr7_bh_June_2010.fits')
print(hdul[1].columns)
datos=Table(hdul[1].data)
```

```python
print(datos)
logl_CIV=datos['LOGL_CIV']
print(logl_CIV)
logE_CIV=datos['LOGL_CIV_ERR']
print(logE_CIV)
```

```python
np.savetxt(path+'LOGL_CIV.csv',logl_CIV, fmt='%5.5f', header='LOGL_CIV')
```

```python
np.savetxt(path+'LOGL_CIV_ERR.csv',logE_CIV, fmt='%5.5f', header='LOGL_CIV')
```
